```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dfINIFAP = pd.read_csv("C://Users/ferna/TESIS/FRESNILLO.csv")

dfINIFAP.head()
```

```python
dfCONAGUA = pd.read_excel("C://Users/ferna/TESIS/TEMPERATURAS_CONAGUA.xlsx")

dfCONAGUA.head()
```

```python
df = pd.DataFrame(data=np.array(dfINIFAP.iloc[:,3]),columns=["TMIN_INIFAP"])

df["TMED_INIFAP"] = np.array(dfINIFAP.iloc[:,4])

df["TMAX_INIFAP"] = np.array(dfINIFAP.iloc[:,2])

df["TMIN_CONAGUA"] = np.array(dfCONAGUA.iloc[:,0])

df["TMED_CONAGUA"] = np.array(dfCONAGUA.iloc[:,1])

df["TMAX_CONAGUA"] = np.array(dfCONAGUA.iloc[:,2])

df
```
